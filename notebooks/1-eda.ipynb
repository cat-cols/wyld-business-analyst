## Project 1: Wyld Ops Command Center

**Goal:** Integrate **Sales + Operations + People** data into one model that answers:

* *What happened?* (performance)
* *Why did it happen?* (drivers)
* *What should we do next?* (actions)

### What you’ll ship (portfolio deliverables)

* Power BI dashboard (3–5 pages)
* SQL model (staging + marts)
* Data dictionary + KPI definitions
* “Reconciliation & Data Quality” page inside Power BI
* 6–8 slide executive deck (screenshots + story)

---

# 1) The data model (star schema)

Keep it simple and “enterprise-looking”:

**Facts**

* `fact_sales` (transactions or daily sales)
* `fact_inventory` (daily inventory snapshots OR stock movements)
* `fact_labor` (labor hours / headcount by date + site/team)

**Dimensions**

* `dim_date`
* `dim_product`
* `dim_location` (store, facility, region)
* `dim_channel` (retail, ecommerce, wholesale)
* `dim_employee_group` (team/department, optional)

This structure screams “I know what I’m doing” and makes DAX easier.

---

# 2) KPIs (pick 12–15, but define them cleanly)

### Sales KPIs

* Revenue
* Units
* Gross Profit
* Gross Margin %
* Average Selling Price (ASP) = Revenue / Units
* Average Order Value (AOV) if you have order-level data

### Ops KPIs

* Days Inventory on Hand (DIO)
* Stockout Rate (days with on_hand = 0 / total days)
* Inventory Turnover (COGS / avg inventory) (optional)
* Fill Rate (if you have demand vs shipped — optional)

### People KPIs

* Labor Hours
* Revenue per Labor Hour
* Overtime % (if available)
* Headcount / Attrition % (optional)

💡💡 Your secret weapon: create a **KPI Definition table** in the repo (`docs/kpi_definitions.md`). Hiring managers love “this person won’t invent nonsense metrics.”

# 3) Dashboard pages (what it should look like)

## Page 1 — Executive Overview

* KPI cards: Revenue, Margin %, DIO, Stockout Rate, Rev/Labor Hr
* Trend line: Revenue & Margin %
* “What changed?” narrative box (more on that below)
* Slicers: Date, Location, Channel, Product Category

## Page 2 — Sales Performance & Drivers

* Revenue and Units trend
* Waterfall: Revenue variance vs prior period (price/volume/mix if possible)
* Top/Bottom products or categories by growth and margin impact

## Page 3 — Inventory & Service Level

* On-hand vs Sales trend
* Stockout heatmap by product/location
* DIO trend + “at risk” items (low days on hand)

## Page 4 — Labor & Efficiency

* Labor hours trend
* Revenue per labor hour trend
* Location/team comparison (who’s efficient vs struggling)

## Page 5 — Data Trust (Reconciliation + QA/QC)

This is what separates you from dashboard hobbyists.

* Source totals vs modeled totals (Revenue, Units, Orders)
* Count of missing keys, duplicates, negative quantities, outlier flags
* “Data quality score” (simple % pass)

---

# 4) “What changed?” narrative (the impressive storytelling piece)

Make a DAX measure that produces plain-English insights, like:

> “Revenue is up 8.2% vs last period, driven mainly by +$42K from Category A and improved margin (+1.4pp). Stockouts increased in 3 locations (Clackamas, Salem, Bend), mostly in SKU group X.”

Even if it’s not perfect AI-level text, basic driver callouts are huge.

Implementation idea:

* Measures for period-over-period change
* Top contributor product/location calculated table or measure
* A concatenated text measure for the “insight” box

# 6) A realistic dataset strategy (so you can actually build it fast)

You have two solid options:

### Option A (fastest): use public data + small synthetic “ops/people”

* Public retail sales dataset for `fact_sales`
* Create `fact_inventory` and `fact_labor` as synthetic but plausible (generated CSVs)
  This is totally acceptable if you’re transparent in the README.

### Option B (more real): use 2–3 public datasets and join on date/location

Harder to find perfect joins, but doable.

Either way: your model + method is what you’re showcasing.
