## 📚 **DAX vs. TABLEAU CALCULATIONS - YOUR CHEAT SHEET**

Since Wyld mentions DAX, here's how to translate DAX concepts to Tableau (so you can speak their language):

| DAX Concept | Tableau Equivalent | Example |
|-------------|-------------------|---------|
| **CALCULATE()** | LOD Expression | `{FIXED [Category] : SUM([Sales])}` |
| **SUMX()** | Row-level calc → AGG | `SUM([Price] * [Quantity])` |
| **RELATED()** | Join/Relationship | Automatic in Tableau relationships |
| **FILTER()** | Context Filter | Use filters or WHERE in calcs |
| **ALL()** | TOTAL() / FIXED | `{FIXED : SUM([Sales])}` |
| **VALUES()** | ATTR() | `ATTR([Field])` |
| **EARLIER()** | LOOKUP() | `LOOKUP(SUM([Sales]), -1)` |
| **RANKX()** | RANK() | `RANK(SUM([Sales]))` |

---

### **Key DAX Patterns → Tableau Translation**

#### **Pattern 1: Calculate with Filter**

**DAX:**
```dax
Oregon Sales =
CALCULATE(
    SUM(Sales[Amount]),
    Sales[State] = "OR"
)
```

**Tableau:**
```
SUM(IF [State] = "OR"
    THEN [Amount]
    END)
```

Or use a filter.

---

#### **Pattern 2: Year-over-Year Growth**

**DAX:**
```dax
YoY Growth =
DIVIDE(
    [Sales],
    CALCULATE([Sales], SAMEPERIODLASTYEAR(Calendar[Date])),
    0
) - 1
```

**Tableau:**
```
(SUM([Sales]) -
 LOOKUP(SUM([Sales]), -12)) /
 LOOKUP(SUM([Sales]), -12)
```

---

#### **Pattern 3: Percentage of Total**

**DAX:**
```dax
Pct of Total =
DIVIDE(
    [Sales],
    CALCULATE([Sales], ALL(Products))
)
```

**Tableau:**
```
SUM([Sales]) /
TOTAL(SUM([Sales]))
```

---

#### **Pattern 4: Cumulative Sum**

**DAX:**
```dax
Cumulative Sales =
CALCULATE(
    [Sales],
    FILTER(
        ALL(Calendar[Date]),
        Calendar[Date] <= MAX(Calendar[Date])
    )
)
```

**Tableau:**
```
RUNNING_SUM(SUM([Sales]))
```

Or use Table Calculation: Running Total
